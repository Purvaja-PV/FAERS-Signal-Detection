```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading our 3 FAERS data files into memory
reac = pd.read_csv(r'C:\Users\hp\Downloads\faers_ascii_2026q1\ASCII\REAC26Q1.txt', sep='$', encoding='latin-1', low_memory=False)
drug = pd.read_csv(r'C:\Users\hp\Downloads\faers_ascii_2026q1\ASCII\DRUG26Q1.txt', sep='$', encoding='latin-1', low_memory=False)
demo = pd.read_csv(r'C:\Users\hp\Downloads\faers_ascii_2026q1\ASCII\DEMO26Q1.txt', sep='$', encoding='latin-1', low_memory=False)

# Confirming data loaded successfully
print("Reactions loaded:", len(reac))
print("Drugs loaded:", len(drug))
print("Patients loaded:", len(demo))
```

```python
# We are choosing ONE drug to investigate
# Let's investigate ACETAMINOPHEN (common painkiller known as Paracetamol in India)
target_drug = 'ACETAMINOPHEN'
target_reaction = 'Death'

# Step 1 - Find all case IDs where our target drug was used
# primaryid is the unique ID number for each patient report
drug_cases = drug[drug['drugname'] == target_drug]['primaryid']
print("Total reports mentioning", target_drug, ":", len(drug_cases))

# Step 2 - Find all case IDs where our target reaction occurred
reaction_cases = reac[reac['pt'] == target_reaction]['primaryid']
print("Total reports mentioning", target_reaction, ":", len(reaction_cases))
```

```python
# Step 3 - Find cases where BOTH the drug AND reaction appear together
# This is called the "observed count" - how many times they appear together
both = len(set(drug_cases) & set(reaction_cases))
print("Reports with BOTH ACETAMINOPHEN and Death:", both)

# Step 4 - Calculate PRR
# A = cases with target drug AND target reaction
A = both

# B = cases with target drug but WITHOUT target reaction  
B = len(drug_cases) - A

# C = cases WITHOUT target drug but WITH target reaction
C = len(reaction_cases) - A

# D = all other cases
D = len(reac) - A - B - C

print("\nOur calculation table:")
print("A (drug + reaction together):", A)
print("B (drug without reaction):", B)
print("C (reaction without drug):", C)
print("D (neither):", D)

# Now calculate PRR
PRR = (A / (A + B)) / (C / (C + D))
print("\nPRR for ACETAMINOPHEN and Death:", round(PRR, 2))

if PRR >= 2:
    print("SIGNAL DETECTED! PRR is above 2 threshold")
else:
    print("No signal detected. PRR is below 2 threshold")
    
```

```python
# Creating a professional summary chart of our PRR finding
# This is what you would present in a real safety meeting

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Chart 1 - Top 10 adverse reactions
# value_counts() counts how many times each reaction appears
# head(10) takes only the top 10
top_reactions = reac['pt'].value_counts().head(10)
axes[0].barh(top_reactions.index, top_reactions.values, color='steelblue')
axes[0].set_title('Top 10 Adverse Reactions\nFAERS 2026 Q1', fontsize=12)
axes[0].set_xlabel('Number of Reports')

# Chart 2 - PRR Signal visualization
# We are showing our PRR value against the safety threshold of 2.0
drugs_to_check = ['ACETAMINOPHEN']
prr_values = [3.29]
colors = ['red' if p >= 2 else 'green' for p in prr_values]
axes[1].bar(drugs_to_check, prr_values, color=colors)
axes[1].axhline(y=2, color='black', linestyle='--', label='Signal Threshold (PRR=2)')
axes[1].set_title('PRR Signal Detection\nACETAMINOPHEN vs Death', fontsize=12)
axes[1].set_ylabel('PRR Value')
axes[1].legend()

plt.tight_layout()
plt.savefig(r'C:\Users\hp\Downloads\faers_signal_dashboard.png')
plt.show()
print("Dashboard saved to Downloads!")
```
